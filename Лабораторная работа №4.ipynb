```python
# бинарный поиск
def binary_search(mas, x):
    # находим среднее значение массива, обозначим верхнюю и нижнюю границу
    middle = len(mas) // 2
    bottom_line = 0
    upper_line = len(mas) - 1
    # сравниваем искомое значение с серединой
    while mas[middle] != x and bottom_line <= upper_line:
        if x > mas[middle]:
            bottom_line = middle + 1
        else:
            upper_line = middle - 1
        # вычисляем индекс новой середины после сдвига одной из границ
        middle = (bottom_line + upper_line) // 2
    # проверка условия, если границы пересеклись
    if bottom_line > upper_line:
        return 'В сгенерированном массиве нет элемента с таким значением'
    else:
        return middle
    
# Вставка элемента
def insert_tree(node, data):
    # Если дерево пустое, то вернуть новый узел
    if node is None:
        return Node(data)
    # Переход в левое/правое поддерево и вставка узла
    if data < node.data:
        node.left = insert_tree(node.left, data)
    else:
        node.right = insert_tree(node.right, data)
    return node

A = []
generate_mas(A)
x = int(input("Введите элемент, который хотите найти: ", ))
starttime = timeit.default_timer()
print(binary_search(A, x))
endtime = timeit.default_timer()
print("Метод поиска работал:", endtime - starttime, "секунд")
a = int(input("Введите элемент, который хотите вставить: ", ))
print("Вывод массива со вставленным элементом: ")
A.append(a)
print(A)
b = int(input("Введите элемент, который хотите удалить: ", ))
print("Вывод массива с удаленным элементом: ")
A.remove(b)
print(A)
```

```python
# Бинарное дерево

# Создание корневого узла
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Сортировка дерева
def inorder(root):
    if root is not None:
        # Обход левого поддерева
        inorder(root.left)
        # Обход корня
        print(str(root.data) + " ", end=' ')
        # Обход правого поддерева
        inorder(root.right)

# Вставка элемента
def insert_tree(node, data):
    # Если дерево пустое, то вернуть новый узел
    if node is None:
        return Node(data)
    # Переход в левое/правое поддерево и вставка узла
    if data < node.data:
        node.left = insert_tree(node.left, data)
    else:
        node.right = insert_tree(node.right, data)
    return node

# Поиск inorder-преемника
def minValueNode(node):
    current = node
    # Inorder-преемник - это найденный крайний левый лист
    while(current.left is not None):
        current = current.left
    return current

# Удаление узла
def deleteNode(root, data):
    # Если дерево пустое, возвращаем
    if root is None:
        return root
    # Необходимо найти узел, который нужно удалить
    if data < root.data:
        root.left = deleteNode(root.left, data)
    elif(data > root.data):
        root.right = deleteNode(root.right, data)
    else:
        # Если у узла только один дочерний узел или вообще их нет
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        # В случае когда у узла два дочерних узла, помещаем центрированного преемника на место узла, который нужно удалить
        temp = minValueNode(root.right)
        root.data = temp.data
        # Удаляем inorder-преемниа
        root.right = deleteNode(root.right, temp.data)
    return root

A = []
generate_mas(A)
root = None
for i in range(len(A)):
    root = insert_tree(root, A[i])
x = int(input("Введите элемент, который хотите найти: ", ))
starttime = timeit.default_timer()
inorder(root)
endtime = timeit.default_timer()
print("\nМетод поиска работал:", endtime - starttime, "секунд")
a = int(input("\nВведите элемент, который хотите вставить: ", ))
root = insert_tree(root, a)
print("\nВывод массива со вставленным элементом: ", end=' ')
inorder(root)
b = int(input("\nВведите элемент, который хотите удалить: ", ))
root = deleteNode(root, b)
print("\nВывод массива с удаленным элементом: ", end=' ')
inorder(root)
```

```python
# Интерполяционный поиск
def interpolation_search(mas, x):
    start_index = 0
    last_index = len(mas)-1
    while start_index <= last_index and x >= mas[start_index] and x <= mas[last_index]:
        # вычисление вероятной позиции искомого элемента
        index = start_index + int(((float(last_index-start_index)/(mas[last_index]-mas[start_index]))*(x-mas[start_index])))
        # элемент найден
        if mas[index] == x:
            return index
        # пересчитываем индекс для праваого подмассива
        elif mas[index] < x:
            start_index = index + 1
        # пересчитываем индекс для левого подмассива
        else:
            last_index = index - 1
    return 'В сгенерированном массиве нет элемента с таким значением'

A = []
generate_mas(A)
x = int(input("Введите элемент, который хотите найти: ", ))
starttime = timeit.default_timer()
print(interpolation_search(A, x))
endtime = timeit.default_timer()
print("Метод поиска работал:", endtime - starttime, "секунд")
```
